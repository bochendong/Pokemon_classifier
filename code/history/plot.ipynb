```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gan_df = pd.read_csv("gan.csv")
baseline_df = pd.read_csv("baseline.csv") 
Mobile_df = pd.read_csv("Mobile.csv") 
Resnet_df = pd.read_csv("Resnet.csv")
```

```python
x = range(0, 30)
gan_train_acc = gan_df['accuracy']
baseline_train_acc = baseline_df['accuracy']
Mobile_train_acc = Mobile_df['accuracy']
Resnet_train_acc = Resnet_df['accuracy']

gan_val_acc = gan_df['val_accuracy']
baseline_val_acc = baseline_df['val_accuracy']
Mobile_val_acc = Mobile_df['val_accuracy']
Resnet_val_acc = Resnet_df['val_accuracy']

gan_train_loss = gan_df['loss']
baseline_train_loss = baseline_df['loss']
Mobile_train_loss = Mobile_df['loss']
Resnet_train_loss = Resnet_df['loss']

gan_val_loss = gan_df['val_loss']
baseline_val_loss = baseline_df['val_loss']

Mobile_val_loss = Mobile_df['val_loss']
Resnet_val_loss = Resnet_df['val_loss']
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15,5),sharey=True)

axs[0].plot(x, baseline_train_acc, 'r')
axs[0].plot(x, baseline_val_acc, 'black')
axs[0].legend(["Trainig", "Val"])
axs[0].title.set_text("Base Model")

axs[1].plot(range(0, 100), Resnet_train_acc, 'r')
axs[1].plot(range(0, 100), Resnet_val_acc, 'black')
axs[1].legend(["Trainig", "Val"])
axs[1].title.set_text("ResNet")

axs[2].plot(range(0, 100), Mobile_train_acc, 'r')
axs[2].plot(range(0, 100), Mobile_val_acc, 'black')
axs[2].legend(["Trainig", "Val"])
axs[2].title.set_text("MobileNet")

```
